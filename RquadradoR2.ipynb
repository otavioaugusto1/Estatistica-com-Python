```python
import pandas as pd
import numpy as np
```

```python
dados = pd.read_csv("https://raw.githubusercontent.com/danmorales/CursoDS_ProfDanilo/master/Estatistica/Calc_R2.csv")
```

```python
dados
```

```python
Y = dados['Y'].values
Y1 = dados['Y1'].values
Y2 = dados['Y2'].values
Y3 = dados['Y3'].values
Y4 = dados['Y4'].values
Y5 = dados['Y5'].values
```

```python
from sklearn.metrics import r2_score #Calcula o R quadrado
```

```python
r2_1 = r2_score(Y,Y1) #linear
r2_2 = r2_score(Y,Y2) #Quadr√°tico
r2_3 = r2_score(Y,Y3) #Assim sucessivamente
r2_4 = r2_score(Y,Y4)
r2_5 = r2_score(Y,Y5)
```

```python
print("Grau 1: ", r2_1)
```

```python
print("R2 (Grau 1) = ",r2_1)
print("R2 (Grau 2) = ",r2_2)
print("R2 (Grau 3) = ",r2_3)
print("R2 (Grau 4) = ",r2_4)
print("R2 (Grau 5) = ",r2_5)
```
